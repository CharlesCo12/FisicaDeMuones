```python
import numpy as np
import matplotlib.pylab as plt
```

```python
datos_Otro=np.loadtxt('Downloads/20-02-28-19-56.data')
datos_Nuest=np.loadtxt('Desktop/20-02-24-11-23.data')
```

```python
i_ot=np.where(datos_Otro[:,0]<40000)
i_n=np.where(datos_Nuest[:,0]<40000)

d1=datos_Otro[i_ot[0],0]
d2=datos_Nuest[i_n[0],0]
d1[d1>14000]=0
d2[d2>14000]=0
```

```python
plt.hist(d1,bins=30,log=True,label='Otro Grupo')
plt.hist(d2,bins=30,log=True,label='Nuestro')
plt.legend()
loc=0.0
```

```python
a_1,b_1=np.histogram(d1,bins=20)
a_2,b_2=np.histogram(d2,bins=20)
```

```python
A_1,B_1=np.polyfit(b_1[1:],np.log(a_1),1)
A_2,B_2=np.polyfit(b_2[1:],np.log(a_2),1)
```

```python
def f_1(x):
    c_1=np.exp(B_1)
    return c_1*np.exp(A_1*x)

def f_2(x):
    c_2=np.exp(B_2)
    return c_2*np.exp(A_2*x)
```

```python
x=np.linspace(0,14000,10000)
plt.plot(x,f_1(x),label='Otro Grupo')
plt.plot(x,f_2(x),label='Nuestro')
plt.legend()
loc=0.0
```

```python
tao_1=abs(1/A_1)/1000
tao_2=abs(1/A_2)/1000

print('El valor de tao para el otro grupo es: ' + str(tao_1))
print('Y su error asociado es de: ' + str(100*abs(tao_1-2.19703)/2.19703) + '%')
print('El valor de tao para nuestro grupo es: ' + str(tao_2))
print('Y su error asociado es de: ' + str(100*abs(tao_2 -2.19703)/2.19703) + '%')
```

```python
datos_extra=np.loadtxt('Downloads/20-01-31-15-35.data')
i_extra=np.where(datos_extra[:,0]<40000)

d3=datos_extra[i_extra[0],0]
d3[d3>14000]=0
```
